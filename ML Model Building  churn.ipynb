```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder,StandardScaler,MinMaxScaler,RobustScaler,LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
df=pd.read_csv("Downloads/churn_dataset.csv")
```

```python
X =df.drop("Churn",axis = 1)
y = df["Churn"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=20)
```

```python
nominal_col = X_train[["gender", "PaperlessBilling","Contract","OnlineSecurity","OnlineBackup","StreamingMovies","MultipleLines" ,"StreamingTV","PhoneService","TechSupport","InternetService", "DeviceProtection","PaymentMethod","InternetService"]]
numerical_col = X_train.select_dtypes(exclude=["object", "category"])
```

```python
ohe = OneHotEncoder(sparse_output=False,drop="first")
```

```python
ohe_array = ohe.fit_transform(nominal_col)
```

```python
ohe_col = ohe.get_feature_names_out()
```

```python
nominal_trans = pd.DataFrame(ohe_array, columns = ohe_col,index= X_train.index)
```

```python
nominal_trans
```

```python
numerical_col 
```

```python
X_train_en = pd.concat([nominal_trans,numerical_col],axis = 1)
```

```python
X_train_en
```

```python
Sc = StandardScaler()
```

```python
X_train_transformed = Sc.fit_transform(X_train_en)
```

```python
X_train_transformed
```

```python
y_train
```

```python
KNN = KNeighborsClassifier()

Model = KNN.fit(X_train_transformed,y_train)
```

```python
X_test
nominal_col = X_test[["gender", "PaperlessBilling","Contract","OnlineSecurity","OnlineBackup","StreamingMovies","MultipleLines" ,"StreamingTV","PhoneService","TechSupport","InternetService", "DeviceProtection","PaymentMethod","InternetService"]]
numerical_col = X_test.select_dtypes(exclude=["object", "category"])
```

```python
ohe_array = ohe.transform(nominal_col)
```

```python
nominal_trans = pd.DataFrame(ohe_array, columns = ohe_col,index= X_test.index)
```

```python
X_test_en = pd.concat([nominal_trans,numerical_col],axis = 1)
```

```python
X_test_transformed = Sc.transform(X_test_en)
```

```python
y_pred = Model.predict(X_test_transformed)
```

```python
accuracy_score(y_test,y_pred)
```

```python
RF = RandomForestClassifier()

Model1 = RF.fit(X_train_transformed,y_train)

y_pred = Model1.predict(X_test_transformed)
```

```python
accuracy_score(y_test,y_pred)
```
